```python
import os
import keras
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions, preprocess_input
from keras.models import Model
```

```python
import numpy as np
import matplotlib.pyplot as plt

def load_image(path):
    img = image.load_img(path, target_size=model.input_shape[1:3])
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return img, x
```

```python
model = keras.applications.VGG16(weights='imagenet', include_top=True)
model.summary()
```

```python
cat, x_cat = load_image('cat.jpg')
dog, x_dog = load_image('dog.jpg')
bus, x_bus = load_image('bus.jpg')
```

```python
plt.imshow(cat)
```

```python
predictions = model.predict(x_cat)
predictions = decode_predictions(predictions)
```

```python
predictions
```

```python
predictions = model.predict(x_dog)
predictions = decode_predictions(predictions)
predictions
```

```python
predictions = model.predict(x_bus)
predictions = decode_predictions(predictions)
predictions
```

```python
feature_extractor = Model(inputs = model.input, outputs=model.get_layer('fc2').output)
feature_extractor.summary()
```

```python
feat_cat = feature_extractor.predict(x_cat)[0]
feat_dog = feature_extractor.predict(x_dog)[0]
feat_bus = feature_extractor.predict(x_bus)[0]
```
