```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
data, labels = iris.data[:, 0:3], iris.data[:,3]

num_samples = len(labels)
shuffle_order = np.random.permutation(num_samples)

data = data[shuffle_order,:]
labels = labels[shuffle_order]

data = data / np.amax(data, axis = 0)
labels = labels / np.amax(labels, axis = 0)
data = data.astype('float32')
labels = labels.astype("float32")
```

```python
print(data.shape)
print(data[0:5,:])
labels[0:5]
```

```python
X, y = data, labels

test_split = 0.3
n_test = int(test_split * num_samples)
x_test, y_test = X[:n_test, :], y[:n_test]
x_train, y_train = X[n_test:, :], y[n_test:]
```

```python
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8,activation='sigmoid', input_dim=3))
model.summary()
```

```python
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
model.fit(x_train, y_train,
         batch_size=4, epochs=200,
         verbose=1,
         validation_data=(x_test,y_test))
```

```python
score = model.evaluate(x_test, y_test)
```

```python
print(score)
```

```python
y_pred = model.predict(x_test)
```

```python
for yp,ya in list(zip(y_pred,y_test))[0:10]:
    print("predicted %0.3f, actual %0.3f"%(yp,ya))
```
